```python
import os
import pandas as pd
import numpy as np
```

```python
foods = pd.read_csv('Food_Preference.csv', header=0)
foods.head()
```

```python
foods = foods.drop(['Timestamp','Juice','Participant_ID'], axis = 1)
foods.head()
```

```python
col = list(foods.columns)
print(foods[col[0]].isnull().values.any())
print(foods[col[1]].isnull().values.any())
print(foods[col[2]].isnull().values.any())
print(foods[col[3]].isnull().values.any())
print(foods[col[4]].isnull().values.any())
```

format age data as float

```python
foods[col[2]] = foods[col[2]].astype('float')
foods[col[3]] = foods[col[3]].astype('str')
foods
```

**we know that in the gender and age colums has nan value so we deal with this nan values**

```python
foods['Gender'].replace(np.nan, "no gender", inplace = True)
ageMedian = foods['Age'].median()
foods['Age'].replace(np.nan, ageMedian, inplace = True)
foods
```

```python
print(foods[col[0]].isnull().values.any())
print(foods[col[1]].isnull().values.any())
print(foods[col[2]].isnull().values.any())
print(foods[col[3]].isnull().values.any())
print(foods[col[4]].isnull().values.any())
```

```python
foods['Food'] = foods['Food'].apply(lambda x: str(x).split(' ')[0])
foods
```
